# Overview

In the Cifar10 dataset there are 10 different categories for the pictures: airplanes, automobiles, birds, cats, deer, dogs, frogs, horses, ships, and trucks. Samples of the photos are provided below. As scroll through the sample photos, note that there is a wide variety of angles, colors, and distances from the object being classified. For example, the photos of planes contain propeller planes, jet engine planes, and stealth fighters; which are all very different looking planes. 

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
(X_train,y_train), (X_test, y_test) = tf.keras.datasets.cifar10.load_data()

# 0 = airplane
# 1 = automobile
# 2 = bird
# 3 = cat
# 4 = deer
# 5 = dog
# 6 = frog
# 7 = horse
# 8 = ship
# 9 = truck
```

```python
airplane = list()
automobile = list()
bird = list()
cat = list()
deer = list()
dog = list()
frog = list()
horse = list()
ship = list()
truck = list()
```

```python
for i,val in enumerate(y_train):
    if val == 0:
        airplane.append(X_train[i])
    if val == 1:
        automobile.append(X_train[i])
    if val == 2:
        bird.append(X_train[i])
    if val == 3:
        cat.append(X_train[i])
    if val == 4:
        deer.append(X_train[i])
    if val == 5:
        dog.append(X_train[i])
    if val == 6:
        frog.append(X_train[i])
    if val == 7:
        horse.append(X_train[i])
    if val == 8:
        ship.append(X_train[i])
    if val == 9:
        truck.append(X_train[i])
```

```python
W_grid = 5
L_grid = 5

fig, axes = plt.subplots(L_grid, W_grid, figsize = (15,15))
axes = axes.ravel()

n_training = len(airplane)

for i in np.arange(0, L_grid * W_grid ):
    index = np.random.randint(0, n_training)
    axes[i].imshow(airplane[index])
    axes[i].set_title('airplane')
    axes[i].axis('off')
```
